# Dividing Dataset between churned and unchurned customers

```python
telecom_churned = telecom_data[telecom_data['Churn'] == True]
telecom_unchurned = telecom_data[telecom_data['Churn'] == False]
```

**Adding column months to classify data into months**

```python
telecom_churned = telecom_churned.assign(Months = lambda x: (x['Account length'] //28))
telecom_unchurned = telecom_unchurned.assign(Months = lambda x: (x['Account length'] //28))
```

```python
# Doing Monthly analysis for both datasets
```

```python
monthy_analysis_churned = telecom_churned.groupby('Months')['Months'].describe()
monthly_analysis_unchurned = telecom_unchurned.groupby('Months')['Months'].describe()
```

**Distribution of churned customers over months**

```python
monthy_analysis_churned.reset_index(inplace=True)
plt.rcParams['figure.figsize'] = (10, 6)
#plt.bar(monthy_analysis_churned['Months'] , monthy_analysis_churned['percentage_customers'])
monthy_analysis_churned = monthy_analysis_churned.assign(Churned_customers_percentage = lambda x: (x['count'] /sum(monthy_analysis_churned['count']) * 100))
sns.barplot(monthy_analysis_churned['Months'] , monthy_analysis_churned['Churned_customers_percentage'], palette = 'hls')
plt.xlabel("Duration(Months)")
plt.ylabel("Percentage of Churned Users")
plt.title("Distribution of churned customers over months")
plt.show()
```

# Above graph signifies that churing became significant after 2 months and between 3-4 months churning was maximum

```python
customer_calls_analysis_churned = telecom_churned.groupby('Customer service calls')['Customer service calls'].describe()
customer_calls_analysis_unchurned = telecom_unchurned.groupby('Customer service calls')['Customer service calls'].describe()
```

```python
telecom_churned.columns
customer_calls_analysis_churned.reset_index(inplace=True)
```

**Distribution of churned customers over customer service calls**

```python
plt.rcParams['figure.figsize'] = (10, 6)
#plt.bar(customer_calls_analysis_churned['Customer service calls'] , customer_calls_analysis_churned['count'])
customer_calls_analysis_churned = customer_calls_analysis_churned.assign(Churned_customers_percentage = lambda x: (x['count'] /sum(monthy_analysis_churned['count']) * 100))
sns.barplot(customer_calls_analysis_churned['Customer service calls'] , customer_calls_analysis_churned['Churned_customers_percentage'], palette = 'hls')
plt.xlabel("Customer Service calls")
plt.ylabel("Percentage of churned users")
plt.title("Distribution of churned customers over customer service calls")
plt.show()
```
